```python
import pandas as pd

# Dataset load
file_path = "/Top Indian Places to Visit.csv"  # Colab e uploaded file er path
df = pd.read_csv(file_path)

# First 5 row dekho
df.head()
```

```python
# Missing values check
print(df.isnull().sum())

# Jodi missing data thake, remove koro
df = df.dropna()
df.reset_index(drop=True, inplace=True)

# Duplicate entries remove koro
df = df.drop_duplicates()
```

```python
from geopy.geocoders import Nominatim

# Geolocator setup
geolocator = Nominatim(user_agent="travel_ranking")

def get_city_coordinates(city_name):
    location = geolocator.geocode(city_name)
    if location:
        return (location.latitude, location.longitude)
    return None
```

```python
from geopy.distance import geodesic

def calculate_distance(user_coords, destination_city):
    dest_coords = get_city_coordinates(destination_city)
    if dest_coords:
        return geodesic(user_coords, dest_coords).km
    return None

```

```python
def rank_destinations(user_city, max_distance=300):
    user_coords = get_city_coordinates(user_city)
    if not user_coords:
        return "City not found!"

    ranked_places = []

    for _, row in df.iterrows():
        destination_city = row['City']  # Dataset e 'City' column ache
        distance = calculate_distance(user_coords, destination_city)

        if distance and distance <= max_distance:
            ranked_places.append((row['Name'], destination_city, distance, row['Google review rating']))

    # Ranking based on distance & Google reviews
    ranked_places.sort(key=lambda x: (x[2], -x[3]))

    return ranked_places[:10]  # Top 10 results return korbe
```

```python
geolocator = Nominatim(user_agent="travel_ranking", timeout=10)


```

```python
import time

def get_city_coordinates(city_name, retries=3):
    for i in range(retries):
        try:
            location = geolocator.geocode(city_name, timeout=10)
            if location:
                return (location.latitude, location.longitude)
        except:
            print(f"Retrying {i+1}/{retries} for {city_name}")
            time.sleep(2)  # Wait 2 seconds before retrying
    return None

```

```python
user_city = "Kolkata"
result = rank_destinations(user_city)

print("Top Weekend Destinations:")
for place in result:
    print(f"{place[0]} in {place[1]} - {round(place[2], 2)} km away (Rating: {place[3]})")

```
